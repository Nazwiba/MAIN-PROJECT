### 7.3  Correcting Data Types

```python
#Gender
df['GENDER'] = df['GENDER'].astype('category')

#Category
df['CATEGORY'] = df['CATEGORY'].astype('category')

#Size
df['SIZE'] = df['SIZE'].astype('category')

#Season
df['SEASON'] = df['SEASON'].astype('category')

#Subscription Status
df['SUBSCRIPTION STATUS'] = df['SUBSCRIPTION STATUS'].map({'Yes': True,'No': False})

#Discount Applied
df['DISCOUNT APPLIED'] = df['DISCOUNT APPLIED'].map({'Yes': True,'No': False})

#Promo code used
df['PROMO CODE USED'] = df['PROMO CODE USED'].map({'Yes':True,'No':False})

#Frequency of Purchase
df['FREQUENCY OF PURCHASES'] = df['FREQUENCY OF PURCHASES'].astype('category')
```

Here converting columns to their most appropriate data types, which:

    Improves memory efficiency
    Speeds up data operations
    Makes the dataset more accurate and consistent

## 9Ô∏è‚É£. VISUALIZATION :

### 9.1. PURCHASE TREND BY GENDER ACROSS CATEGORIES 
LINEPLOT

```python
purchase_trend = df.groupby(['GENDER', 'CATEGORY'])['PURCHASE AMOUNT (USD)'].sum().reset_index()

plt.figure(figsize=(8,6))
sns.lineplot(
    data=purchase_trend,
    x='CATEGORY',
    y='PURCHASE AMOUNT (USD)',
    hue='GENDER',
    marker='o'
)
plt.title("PURCHASE TREND BY GENDER ACROSS CATEGORY", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Avg Purchase Amount (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
purchase_trend
```

#### This Line Plot shows :-
    Males spend more than females across all categories.
    Clothing is the highest-spending category for both genders.
    Spending drops sharply for Footwear and even more for Outerwear.
    Female spending is more stable, while male spending shows stronger peaks and drops.
    Category preferences differ by gender, suggesting the need for targeted product and marketing strategies.

### 9.2. AVERAGE PURCHASE AMOUNT BY AGE GROUP
LINEPLOT

```python
age_order = ['Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']

# Calculate average purchase amount for each age group
age_group_purchase = (
    df.groupby('AGE GROUP')['PURCHASE AMOUNT (USD)'].mean()
      .reindex(age_order)
      .reset_index()
)
plt.figure(figsize=(10,6))
sns.lineplot(data=age_group_purchase, x='AGE GROUP', y='PURCHASE AMOUNT (USD)', marker='o')
plt.title('AVERAGE PURCHASE AMOUNT BY AGE GROUP',fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Average Purchase Amount (USD)')
plt.grid(True)
plt.show()
age_group_purchase
```

#### This Line Plot show:-
    Teens show the widest spread in spending, indicating inconsistent purchase behavior.
    Young Adults have moderate spending with some high-value outliers.
    Adults maintain stable, mid-range purchase amounts with fewer extremes.
    Middle Age customers show controlled, consistent spending with narrow variability.
    Seniors generally spend less with very low variation in purchase amounts.

### 9.3. PURCHASE AMOUNT DISTRIBUTION BY GENDER
PIE CHART

```python
# Calculate total purchase amount for each gender
gender_purchase = df.groupby('GENDER')['PURCHASE AMOUNT (USD)'].sum()

plt.figure(figsize=(12,8))
plt.pie(
    gender_purchase,
    autopct='%1.1f%%',
)
plt.legend(gender_purchase.index, title="Gender")
plt.title('PURCHASE AMOUNT DISTRIBUTION BY GENDER',fontsize=14)
plt.show()
gender_purchase
```

#### This Pie Chart show:-
    Male customers dominate overall spending, contributing nearly twice as much as female customers.
    The 67.7% share indicates stronger purchase intensity or higher-value transactions among male customers.
    The 32.3% share from females suggests either fewer purchases or lower average purchase amounts.
    The imbalance hints that marketing strategies targeting male shoppers may currently be more effective.
    It also suggests an opportunity to develop campaigns to increase engagement and spending among female customers.



### 9.4. AVERAGE PURCHASE AMOUNT BY SEASON 
BARPLOT

```python
# Calculate average purchase amount for each season
season_purchase = df.groupby('SEASON')['PURCHASE AMOUNT (USD)'].mean().reset_index()
colors = ["#F34848", "#207AD3", "#27D527", "#DD8A37"]
plt.figure(figsize=(10,6))
sns.barplot(data=season_purchase, x='SEASON', y='PURCHASE AMOUNT (USD)',palette=colors)

plt.title('AVERAGE PURCHASE AMOUNT BY SEASON',fontsize=14)
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
season_purchase
```

#### This Bar Plot shows:-
    Fall has the highest average purchase amount, indicating stronger buying behavior during this season.
    Spring and Summer show slightly lower purchase averages, suggesting relatively reduced spending in these periods.
    Winter purchases are moderately high, close to Fall, showing steady consumer demand during colder months.
    Overall, seasonal variation is small, indicating consistent spending across all seasons with only slight peaks.



### 9.5. ITEM PURCHASED BY SEASON AND CATEGORY
COUNTPLOT

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='SEASON', hue='CATEGORY')

plt.title("ITEMS PURCHASED BY SEASON AND CATEGORY",fontsize=14)
plt.xlabel("Season")
plt.ylabel("Number of Items Purchased")
plt.legend(title="Category")
plt.show()
```

    Footwear demand peaks in Spring and Summer, slightly dropping in Fall and Winter.

### 9.6. TOP 10 LOCATION - AVERAGE PURCHASE AMOUNT BY CATEGORY
HEATMAP

```python
# Top 10 locations
top_locations = (
    df.groupby('LOCATION')['PURCHASE AMOUNT (USD)']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

df_top = df[df['LOCATION'].isin(top_locations)]

pivot = df_top.pivot_table(
    values='PURCHASE AMOUNT (USD)',
    index='LOCATION',
    columns='CATEGORY',
    aggfunc='mean'
)
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, cmap='Blues', fmt=".1f")

plt.title('TOP 10 LOCATIONS - AVERAGE PURCHASE AMOUNT BY CATEGORY',fontsize=14)
plt.xlabel('Category')
plt.ylabel('Location')
plt.show()
top_locations

```

#### This HeatMap shows:-
    Illinois shows the highest purchase amounts, especially for Outerwear (73.8) and Accessories (68.7).
    Nevada has strong spending in Accessories and Clothing, but very low Outerwear spending (31.2).
    Idaho, Nebraska, and North Dakota show consistent above-average spending, especially in Outerwear.
    Montana has the lowest Footwear spending (46.2).
    Overall, Outerwear spending varies the most, while Clothing shows stable demand across locations.

### 9.9. PURCHASE AMOUNT DISTRIBUTION BY SUBSCRIPTION STATUS
PIE CHART

```python
# Aggregate purchase amount by subscription status
sub_amount = df.groupby("SUBSCRIPTION STATUS")["PURCHASE AMOUNT (USD)"].sum()

plt.figure(figsize=(8,8))
plt.pie(sub_amount, 
        labels=sub_amount.index, 
        autopct='%1.1f%%', 
        )
plt.title("PURCHASE AMOUNT DISTRIBUTION BY SUBSCRIPTION STATUS",fontsize=14)
plt.legend(title="Subscription Status")
plt.show()
sub_amount
```

#### This Pie Chart shows:-
    A large majority (73.1%) of the total purchase amount comes from non-subscribed customers.
    Only 26.9% of purchase amount comes from subscribed customers.
    This indicates that most revenue is driven by customers who are not subscribed
    The business may need to improve subscription value or run engagement campaigns to increase purchases among subscribers.

### 9.10. PURCHASE AMOUNT BY CATEGORY WITH DISCOUNT APPLIED
BOXPLOT

```python
plt.figure(figsize=(8,6))
sns.boxplot(
    data=df,
    x="CATEGORY",
    y="PURCHASE AMOUNT (USD)",
    hue="DISCOUNT APPLIED"
)

plt.title("PURCHASE AMOUNT BY CATEGORY WITH DISCOUNT APPLIED",fontsize=14)
plt.xlabel("Category")
plt.ylabel("Purchase Amount (USD)")
plt.legend(title="Discount Applied")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### This Box plot shows:-
    Purchase amounts are similar across all categories whether a discount is applied or not‚Äîno major shift in medians.
    Discounted purchases show slightly lower median in Outerwear but similar in other categories.
    Variability (spread) is almost identical for ‚ÄúYes‚Äù and ‚ÄúNo‚Äù across all categories.
    No category shows a strong discount-driven change in spending behavior.
    Overall: discounts do not significantly change purchase amounts in any category.
